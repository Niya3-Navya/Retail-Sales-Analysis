```python
# import pandas as pd
# import numpy as np
# import matplotlib.pyplot as plt
# import seaborn as sns
# import sqlite3
# import warnings


# display(region_sales)
# warnings.filterwarnings('ignore')

# # Load data
# df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
# print("Shape:", df.shape)
# print(df.head())

# # Load into SQLite
# conn = sqlite3.connect('superstore.db')
# df.to_sql('sales', conn, if_exists='replace', index=False)

# # SQL Query 1 - Region wise sales
# region_sales = pd.read_sql_query("""
#     SELECT Region, 
#     ROUND(SUM(Sales),2) as Total_Sales,
#     ROUND(SUM(Profit),2) as Total_Profit
#     FROM sales GROUP BY Region 
#     ORDER BY Total_Sales DESC
# """, conn)
# print("\nRegion wise Sales:\n", region_sales)
# from IPython.display import display

# # SQL Query 2 - Category wise profit
# category = pd.read_sql_query("""
#     SELECT Category,
#     ROUND(SUM(Sales),2) as Total_Sales,
#     ROUND(SUM(Profit),2) as Total_Profit
#     FROM sales GROUP BY Category
#     ORDER BY Total_Profit DESC
# """, conn)
# print("\nCategory wise Profit:\n", category)

# # SQL Query 3 - Top 10 products
# top_products = pd.read_sql_query("""
#     SELECT `Product Name`,
#     ROUND(SUM(Sales),2) as Total_Sales
#     FROM sales GROUP BY `Product Name`
#     ORDER BY Total_Sales DESC LIMIT 10
# """, conn)
# print("\nTop 10 Products:\n", top_products)

# # SQL Query 4 - Segment wise
# segment = pd.read_sql_query("""
#     SELECT Segment,
#     COUNT(*) as Total_Orders,
#     ROUND(SUM(Sales),2) as Total_Sales
#     FROM sales GROUP BY Segment
# """, conn)
# print("\nSegment wise:\n", segment)

# # SQL Query 5 - Disco



# display(category)
# display(top_products)
# display(segment)

# #Region-wise Sales Chart
# plt.figure(figsize=(8,5))
# sns.barplot(x='Region', y='Total_Sales', data=region_sales)
# plt.title("Region Wise Sales")
# plt.xticks(rotation=45)
# plt.show()


# #Category-wise Profit
# plt.figure(figsize=(6,4))
# sns.barplot(x='Category', y='Total_Profit', data=category)
# plt.title("Category Wise Profit")
# plt.show()


# #Top 10 Products
# plt.figure(figsize=(10,6))
# sns.barplot(y='Product Name', x='Total_Sales', data=top_products)
# plt.title("Top 10 Products by Sales")
# plt.show()


# #Segment-wise Sales
# plt.figure(figsize=(6,4))
# sns.barplot(x='Segment', y='Total_Sales', data=segment)
# plt.title("Segment Wise Sales")
# plt.show()


# #Make it look like a Dashboard
# print("📊 SUPERSTORE SALES DASHBOARD")

# display(region_sales)
# display(category)

# plt.figure(figsize=(12,5))

# plt.subplot(1,2,1)
# sns.barplot(x='Region', y='Total_Sales', data=region_sales)
# plt.title("Region Sales")

# plt.subplot(1,2,2)
# sns.barplot(x='Category', y='Total_Profit', data=category)
# plt.title("Category Profit")

# plt.tight_layout()
# plt.show()
















import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import warnings
from IPython.display import display
warnings.filterwarnings('ignore')

# Load data
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
print("Shape:", df.shape)

# Load into SQLite
conn = sqlite3.connect('superstore.db')
df.to_sql('sales', conn, if_exists='replace', index=False)

# SQL Query 1 - Region wise sales
region_sales = pd.read_sql_query("""
    SELECT Region, 
    ROUND(SUM(Sales),2) as Total_Sales,
    ROUND(SUM(Profit),2) as Total_Profit
    FROM sales GROUP BY Region 
    ORDER BY Total_Sales DESC
""", conn)

# SQL Query 2 - Category wise profit
category = pd.read_sql_query("""
    SELECT Category,
    ROUND(SUM(Sales),2) as Total_Sales,
    ROUND(SUM(Profit),2) as Total_Profit
    FROM sales GROUP BY Category
    ORDER BY Total_Profit DESC
""", conn)

# SQL Query 3 - Top 10 products
top_products = pd.read_sql_query("""
    SELECT [Product Name],
    ROUND(SUM(Sales),2) as Total_Sales
    FROM sales GROUP BY [Product Name]
    ORDER BY Total_Sales DESC LIMIT 10
""", conn)

# SQL Query 4 - Segment wise
segment = pd.read_sql_query("""
    SELECT Segment,
    COUNT(*) as Total_Orders,
    ROUND(SUM(Sales),2) as Total_Sales
    FROM sales GROUP BY Segment
""", conn)

print("✅ All SQL queries done!")
display(region_sales)
display(category)
display(top_products)
display(segment)

# FINAL DASHBOARD
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Superstore Retail Sales Dashboard', fontsize=18, fontweight='bold')

# Chart 1 - Region Sales
axes[0,0].bar(region_sales['Region'], region_sales['Total_Sales'], 
              color=['#2ecc71','#3498db','#e74c3c','#f39c12'])
axes[0,0].set_title('Sales by Region', fontweight='bold')
axes[0,0].set_ylabel('Total Sales ($)')

# Chart 2 - Category Profit
axes[0,1].bar(category['Category'], category['Total_Profit'],
              color=['#9b59b6','#1abc9c','#e67e22'])
axes[0,1].set_title('Profit by Category', fontweight='bold')
axes[0,1].set_ylabel('Total Profit ($)')

# Chart 3 - Segment Pie
axes[0,2].pie(segment['Total_Sales'], labels=segment['Segment'], 
              autopct='%1.1f%%', colors=['#3498db','#e74c3c','#2ecc71'],
              startangle=90)
axes[0,2].set_title('Sales by Customer Segment', fontweight='bold')

# Chart 4 - Top 10 Products
axes[1,0].barh(top_products['Product Name'].str[:25], 
               top_products['Total_Sales'], color='#3498db')
axes[1,0].set_title('Top 10 Products by Sales', fontweight='bold')
axes[1,0].set_xlabel('Total Sales ($)')

# Chart 5 - Monthly Sales Trend
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.to_period('M')
monthly = df.groupby('Month')['Sales'].sum().reset_index()
monthly['Month'] = monthly['Month'].astype(str)
axes[1,1].plot(range(len(monthly)), monthly['Sales'], 
               color='#2ecc71', linewidth=2, marker='o', markersize=3)
axes[1,1].set_title('Monthly Sales Trend', fontweight='bold')
axes[1,1].set_ylabel('Sales ($)')
axes[1,1].set_xticks(range(0, len(monthly), 6))
axes[1,1].set_xticklabels(monthly['Month'][::6], rotation=45)

# Chart 6 - Region Profit
axes[1,2].bar(region_sales['Region'], region_sales['Total_Profit'],
              color=['#e74c3c','#f39c12','#2ecc71','#3498db'])
axes[1,2].set_title('Profit by Region', fontweight='bold')
axes[1,2].set_ylabel('Total Profit ($)')

plt.tight_layout()
plt.savefig('dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Dashboard saved as dashboard.png!")
```
